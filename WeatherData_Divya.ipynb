```python
variables = ['Temperature AVG', 'Relative Humidity AVG',
       'Solare Radiation AVG', 'Wind Speed Daily AVG', 'Wind Direction AVG']
from scipy import stats
z_scores = stats.zscore(data['Temperature AVG'])
z_score_outliers = (z_scores < -3) | (z_scores > 3)
z_score_outlier_rows = data[z_score_outliers]
print("Outliers Detected by Z-Score:")
print(z_score_outlier_rows)
threshold = 3
dno1 = data[(z_scores < threshold) & (z_scores > -threshold)]
```

```python
from scipy import stats
z_scores = stats.zscore(dno1['Relative Humidity AVG'])
z_score_outliers = (z_scores < -3) | (z_scores > 3)
z_score_outlier_rows = dno1[z_score_outliers]
print("Outliers Detected by Z-Score:")
print(z_score_outlier_rows)
threshold = 3
dno2 = dno1[(z_scores < threshold) & (z_scores > -threshold)]
```

```python
from scipy import stats
z_scores = stats.zscore(dno2['Relative Humidity AVG'])
z_score_outliers = (z_scores < -3) | (z_scores > 3)
z_score_outlier_rows = dno2[z_score_outliers]
print("Outliers Detected by Z-Score:")
print(z_score_outlier_rows)
threshold = 3
dno3 = dno2[(z_scores < threshold) & (z_scores > -threshold)]
```

```python
from scipy import stats
z_scores = stats.zscore(dno3['Wind Speed Daily AVG'])
z_score_outliers = (z_scores < -3) | (z_scores > 3)
z_score_outlier_rows = dno3[z_score_outliers]
print("Outliers Detected by Z-Score:")
print(z_score_outlier_rows)
threshold = 3
dno4 = dno3[(z_scores < threshold) & (z_scores > -threshold)]
```

```python
from scipy import stats
z_scores = stats.zscore(dno4['Wind Direction AVG'])
z_score_outliers = (z_scores < -3) | (z_scores > 3)
z_score_outlier_rows = dno4[z_score_outliers]
print("Outliers Detected by Z-Score:")
print(z_score_outlier_rows)
threshold = 3
df = dno4[(z_scores < threshold) & (z_scores > -threshold)]
```

```python
df
```

```python
df=df.drop(['Date'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler().fit(df)
df = scale.transform(df)
features_scaled = pd.DataFrame( df, columns= data.columns[1:6])
features_scaled.head()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,6):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(features_scaled)
    
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
sns.set()
plt.plot(range(1,6),wcss)
plt.title('Elbow curve')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=2,init='k-means++',random_state=0)
y=kmeans.fit_predict(features_scaled)
```

```python
print(y)
```

```python
data_output = features_scaled.copy(deep = True)
data_output['Cluster'] = kmeans.labels_
data_output.head()
```

```python
sns.countplot(x='Cluster',data=data_output)
plt.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score,davies_bouldin_score
```

```python
silhouette = silhouette_score(features_scaled, y)
print(f"Silhouette Score: {silhouette}")
```

```python
calinski_harabasz_index = calinski_harabasz_score(features_scaled, y)
print(f"Calinski-Harabasz Index: {calinski_harabasz_index}")
```

```python
davies_bouldin_index = davies_bouldin_score(features_scaled, y)
print(f"Davies-Bouldin Index: {davies_bouldin_index}")
```

```python

import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import scale as s
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z = sch.linkage(features_scaled,method='ward')
Z
```

```python
den = sch.dendrogram(Z)
plt.tick_params(
    axis='x',          
    which='both',      
    bottom=False,     
    top=False,         
    labelbottom=False) 
plt.title('Hierarchical Clustering')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc_model = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage ='ward')
```

```python
y_cluster = hc_model.fit_predict(features_scaled)
```
